```python
import os
from flask import Flask, flash, request, redirect, url_for, jsonify, send_file, make_response
# from werkzeug.utils import secure_filename
import easyocr
from flask_cors import CORS, cross_origin
import urllib
import cv2
import base64
import numpy as np
import matplotlib.pyplot as plt
from pdf2image import convert_from_path, convert_from_bytes
from docx import Document
from docx.enum.style import WD_STYLE_TYPE as ST
from docx.enum.text import WD_ALIGN_PARAGRAPH, WD_COLOR_INDEX
from docx.shared import Pt, RGBColor
import io
import smtplib
import ssl
from email.message import EmailMessage

def detect_text(file, network_dir):
    
    reader = easyocr.Reader(['ur'], user_network_directory = network_dir)
    result = reader.readtext(file, detail=1)
    return result

def image_box_maker(file, file_type, box_coordinates):
    if file_type == "byte":
        image = np.asarray(bytearray(file), dtype="uint8")
        img = cv2.imdecode(image, flags=cv2.IMREAD_UNCHANGED)
    else:
        img = np.asarray(file)
    
    img = cv2.bitwise_not(img)
    
    lst = []
    for i in box_coordinates:
        sub_lst = []
        sub_lst.append((int(i[0][0][0]), int(i[0][0][1])))
        sub_lst.append((int(i[0][2][0]), int(i[0][2][1])))
        lst.append(sub_lst)
    print(lst)
    for i in lst:
        cv2.rectangle(img, i[0], i[1], (0, 200, 255), 2)
    
    return img



def result_join(result):
    text = ""
    for i in range(len(result)-1, -1, -1):
        text+= result[i][1]
    return text

def encode_files(img):

    retval, buffer = cv2.imencode('.jpg', img)
    jpg_as_text = base64.b64encode(buffer).decode()
    return jpg_as_text

def convert_to_images(file, poppler=r"C:\Working_Directory\Internship_NCL_MainWorking\Urdu legatures\poppler-0.68.0\bin"):
    images = convert_from_bytes(file, poppler_path=poppler)
    return images


def mail_send(login_mail, login_pass, send_mail, rec_mail, heading, message):
    em = EmailMessage()
    em['From'] = send_mail
    em['To'] = rec_mail
    em['Subject'] = heading
    em.set_content(message)

    context = ssl.create_default_context()
    with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
        smtp.login(login_mail, login_pass)
        smtp.sendmail(send_mail, rec_mail, em.as_string())

    



# UPLOAD_FOLDER = 'D:\All Downloads'
ALLOWED_EXTENSIONS = {'txt', 'pdf', 'png', 'jpg', 'jpeg', 'gif'}

app = Flask(__name__)
CORS(app, support_credentials=True)
app.config['CORS_HEADERS'] = 'Content-Type'
# app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

# def allowed_file(filename):
#     return '.' in filename and \
#            filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS

@app.route('/', methods=['GET', 'POST'])
@cross_origin(supports_credentials=True)
def upload_file():
    if request.method == 'POST':
        # check if the post request has the file part
#         print(request.files)
#         if 'file' not in request.files:
#             flash('No file part')
#             return redirect(request.url)
        file = request.get_json()
        # If the user does not select a file, the browser submits an
        # empty file without a filename.
#         if file.filename == '':
#             flash('No selected file')
#             return redirect(request.url)
#         if file and allowed_file(file.filename):
#             print(file)
#             filename = secure_filename(file.filename)
#             print(filename)
#             file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
#             print(os.path.join(app.config['UPLOAD_FOLDER'], filename))
#             print(f'D:/All Downloads/{filename}')
        f = urllib.request.urlopen(file["file"])
        myfile = f.read()
        if f.info()['content-type'].split("/")[0] == "image":
            
            result_ocr = detect_text(myfile, 'C:/Users/Syed Muhammad/.EasyOCR/user_network/lig')
            final_result_text = result_join(result_ocr)
            boxed_image = image_box_maker(myfile, "byte", result_ocr)
            encoded_boxed_image = encode_files(boxed_image)
            return jsonify({
                "text": final_result_text,
                "image": encoded_boxed_image
            })
        
        else:
            
            file_images = convert_to_images(myfile)
            if len(file_images)>3:
                file_images = file_images[:3]
            final_text = ""
            final_image = None
            for image in file_images:
                result_ocr = detect_text(np.asarray(image), 'C:/Users/Syed Muhammad/.EasyOCR/user_network/lig')
                final_text += result_join(result_ocr)
                print(result_ocr)
                boxed_image = image_box_maker(image, "not_byte", result_ocr)
                if final_image == None:
                    final_image = boxed_image
                else:
                    final_image = np.concatenate((final_image, boxed_image), axis=0)
                
            encoded_boxed_image = encode_files(final_image)
            return jsonify({
                "text": final_text,
                "image": encoded_boxed_image
            })
                

@app.route('/download', methods=['POST'])
@cross_origin(supports_credentials=True)           
def download_file():
    text = request.get_json()["text"]
    
    document = Document()
    styles = document.styles
    style_A = styles.add_style("A", ST.CHARACTER)
    style_A.font.name = "Urdu Typesetting"
    style_A.font.bold= False
    style_A.font.size = Pt(30)
    
    paragraph = document.add_paragraph()
    para = paragraph.add_run(text, style_A)
#     para.font.highlight_color=WD_COLOR_INDEX.BLACK
    para.font. color. rgb = RGBColor(0x00, 0x00, 0x00)
    bio = io.BytesIO()
    document.save(bio)  
    bio.seek(0)
    print(bio.getvalue())
    return send_file(bio, mimetype="application/vnd.openxmlformats-officedocument.wordprocessingml.document")
   
  


@app.route('/send_mail', methods=['POST'])
@cross_origin(supports_credentials=True)           
def sending_mail():
    f_name = request.get_json()["first_name"]
    l_name = request.get_json()["last_name"]
    sender_email = request.get_json()["mail"]
    message = request.get_json()["mess"]
    
    login_email = "syedmuhammad1111@gmail.com"
    login_pass = "ibunimyxdgejppex"
    
    title = f"It's the message from {f_name} {l_name}"
    print("hmmm")
    mail_send(login_email, login_pass, sender_email, "syedmuhammad0123@gmail.com", title, message)
    return "sent"
            
#             reader = easyocr.Reader(['ur'], user_network_directory = 'C:/Users/Syed Muhammad/.EasyOCR/user_network/lig')
#             result = reader.readtext(myfile, detail=1)
# #         print("file:   ", myfile)
# #         with open("./test.png", 'wb') as file_handler:
# #             file_handler.write(myfile)
        
        
        
#         print(result)
#         image = np.asarray(bytearray(myfile), dtype="uint8")
#         img = cv2.imdecode(image, flags=cv2.IMREAD_UNCHANGED)
#         img = cv2.bitwise_not(img)
# #         print(img)
#         lst = []
#         for i in result:
#             sub_lst = []
#             sub_lst.append((i[0][0][0], i[0][0][1]))
#             sub_lst.append((i[0][2][0], i[0][2][1]))
#             lst.append(sub_lst)
        
#         for i in lst:
#             cv2.rectangle(img, i[0], i[1], (0, 200, 255), 2)
            
#         text = ""
#         for i in range(len(result)-1, -1, -1):
#             text+= result[i][1]
            
#         retval, buffer = cv2.imencode('.jpg', img)
#         jpg_as_text = base64.b64encode(buffer).decode()
# #         string = base64.b64encode(cv2.imencode('.png', img)[1]).decode()
        
# #         response = make_response(jsonify(send_file(jpg_as_text, mimetype='image/png')), 00)
# #         response.headers['Content-Transfer-Encoding']='base64'
#         print(text)
# #         os.remove("./test.png")
        
    
        
#         return jsonify({
        
#         "text" : text, 
#         "image" : jpg_as_text
            
        
#         })

app.run(debug=False)
```
